```python
!pip install ipywidgets
import ipywidgets as widgets
from IPython.display import display, clear_output

# Initialize the board
board = [' '] * 9
current_player = 'X'

def check_winner():
    for i in range(0, 9, 3):
        if board[i] == board[i + 1] == board[i + 2] != ' ':
            return board[i]

    for i in range(3):
        if board[i] == board[i + 3] == board[i + 6] != ' ':
            return board[i]

    if board[0] == board[4] == board[8] != ' ':
        return board[0]
    if board[2] == board[4] == board[6] != ' ':
        return board[2]

    return None

def button_click(b):
    global current_player
    index = int(b.description)
    if board[index] == ' ':
        board[index] = current_player
        b.style.button_color = 'lightblue' if current_player == 'X' else 'lightgreen'
        b.label = current_player
        winner = check_winner()
        if winner:
            status.value = f"Player {winner} wins!"
            for button in buttons:
                button.disabled = True
        elif ' ' not in board:
            status.value = "It's a tie!"
        else:
            current_player = 'O' if current_player == 'X' else 'X'
            status.value = f"Player {current_player}'s turn"
    update_board()

def update_board():
    for i in range(9):
        buttons[i].description = board[i]
        buttons[i].disabled = board[i] != ' '

buttons = [widgets.Button(description=str(i), layout=widgets.Layout(width='80px', height='80px')) for i in range(9)]
for button in buttons:
    button.on_click(button_click)

grid = widgets.GridBox(buttons, layout=widgets.Layout(grid_template_columns="repeat(3, 80px)"))
status = widgets.Label(value=f"Player {current_player}'s turn")

display(grid, status)

def reset_game(b):
    global board, current_player
    board = [' '] * 9
    current_player = 'X'
    status.value = f"Player {current_player}'s turn"
    for button in buttons:
        button.style.button_color = ''
        button.disabled = False
    update_board()

reset_button = widgets.Button(description="Reset", button_style='warning')
reset_button.on_click(reset_game)

display(reset_button)
```
